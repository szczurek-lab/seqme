```python
from itertools import combinations
from collections import Counter
from typing import List
from pepme.utils import read_fasta_file
import pandas as pd

def subsequences(s: str, k: int):
    return [s[i : i + k] for i in range(len(s) - k + 1)]

amp_path = ".././data/dbaasp.fasta"
uniprot_path = ".././data/uniprot/uniprot_8_50_100.fasta"
veltri_path = ".././data/veltri_positive.csv"
dramp_path = ".././data/amp_positive.fasta"

amp_data = read_fasta_file(amp_path)
uniprot_data = read_fasta_file(uniprot_path)
dramp_data = read_fasta_file(dramp_path)
veltri_data = pd.read_csv(veltri_path)["Sequence"]



datasets = {"dramp": dramp_data, "veltri": veltri_data, "dbaasp": amp_data}

for dataset, data in datasets.items():
    
    aggregated_count = Counter()
    
    for seq in data:
        aggregated_count += Counter(subsequences(seq, k=3))
        
    print(f"{dataset} \n\n {aggregated_count} \n")

```
