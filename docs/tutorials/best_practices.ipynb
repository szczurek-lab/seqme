# Best Practices

In this notebook, we will curate papers suggesting metrics for each domain.

## Small molecules

- [Fréchet ChemNet Distance: A metric for generative models for molecules in drug discovery (2018)](https://arxiv.org/abs/1803.09518)
- [MOSES: A Benchmarking Platform for Molecular Generation Models (2020)](https://arxiv.org/abs/1811.12823)
- [GuacaMol: Benchmarking Models for De Novo Molecular Design (2019)](https://arxiv.org/abs/1811.09621)
- [How Evaluation Choices Distort the Outcome of Generative Drug Discovery (2025)](https://arxiv.org/abs/2501.05457)

## DNA

- [DiscDiff: Latent Diffusion Model for DNA Sequence Generation (2024)](https://arxiv.org/abs/2402.06079) — introduces Fréchet Reconstruction Distance (FReD), a DNA analog to FID

## RNA

- [RNAInvBench: Benchmark for the RNA Inverse Design Problem (2024)](https://openreview.net/forum?id=zmzBD74UXj)
- [Deep Generative Models Design mRNA Sequences with Enhanced Translational Capacity and Stability (2025)](https://www.science.org/doi/10.1126/science.adr8470)

## Peptides

No standardized benchmark platform exists yet (analogous to MOSES for small molecules) — current papers use ad hoc combinations of perplexity, novelty, diversity, and property predictors.

- [Deep Generative Models for Peptide Design (2022)](https://pubs.rsc.org/en/content/articlelanding/2022/dd/d1dd00024a) — review explicitly noting the lack of standardized benchmarks
- [Deep Generative Models for Therapeutic Peptide Discovery: A Comprehensive Review (2025)](https://dl.acm.org/doi/10.1145/3714455)

## Proteins

- [Towards Robust Evaluation of Protein Generative Models: A Systematic Analysis of Metrics (2025)](https://www.biorxiv.org/content/10.1101/2024.10.25.620213v1)
- [Protein FID: Improved Evaluation of Protein Structure Generative Models (2025)](https://arxiv.org/abs/2505.08041)
- [ProteinBench: A Holistic Evaluation of Protein Foundation Models (2024)](https://arxiv.org/abs/2409.06744)
