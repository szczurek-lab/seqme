# Visualization


In this notebook, we show some of the plotting functions for property models and embedding models.


```python
import matplotlib.pyplot as plt
import pandas as pd

from pepme.models.embeddings import PropertyEmbedding
from pepme.models.embeddings.plots import plot_pca, plot_tsne, plot_umap
from pepme.models.normalizers import MinMaxNorm
from pepme.models.properties import Charge, Gravy, MolecularWeight
from pepme.models.properties.plots import hist, kde, violin
from pepme.utils import random_subset, read_fasta_file
```

## Property visualization


Let's start by loading some sequences.


```python
DATASET_PATH = "./data/amp-diffusion.fasta"
```

```python
sequences = read_fasta_file(DATASET_PATH)
```

Let's create a histogram, kde and violin plot for a few properties.


```python
properties = {"Gravy": Gravy(), "Molecular Weight": MolecularWeight()}
```

```python
df = pd.DataFrame({name: prop(sequences) for name, prop in properties.items()})
```

```python
per_ax_size = (4, 3)
n = len(properties)
fig, axes = plt.subplots(ncols=n, figsize=(per_ax_size[0] * n, per_ax_size[1]))

for i in range(len(axes)):
    col_name = df.columns[i]
    col = df[col_name]
    hist(col.to_numpy(), xlabel=col_name, ax=axes[i])

fig.tight_layout()
```

```python
per_ax_size = (4, 3)
n = len(properties)
fig, axes = plt.subplots(ncols=n, figsize=(per_ax_size[0] * n, per_ax_size[1]))

for i in range(len(axes)):
    col_name = df.columns[i]
    col = df[col_name]
    kde(col.to_numpy(), xlabel=col_name, ax=axes[i])

fig.tight_layout()
```

```python
per_ax_size = (4, 3)
n = len(properties)
fig, axes = plt.subplots(ncols=n, figsize=(per_ax_size[0] * n, per_ax_size[1]))

for i in range(len(axes)):
    col_name = df.columns[i]
    col = df[col_name]
    violin(col.to_numpy(), xlabel=col_name, ax=axes[i])

fig.tight_layout()
```

## Embedding visualization


Let's start by loading some sequences.


```python
DATASET_PATHS = {
    "CPL-Diff": "./data/cpl-diff.fasta",
    "OmegAMP": "./data/omegamp.fasta",
    "DBAASP": "./data/dbaasp.fasta",
    "UniProt": "./data/uniprot/uniprot_8_50_100.fasta",
}
```

```python
datasets = {name: read_fasta_file(path) for name, path in DATASET_PATHS.items()}
```

```python
for model_name, sequences in datasets.items():
    print(f"{model_name}: {len(sequences)} sequences")
```

```python
n_samples = 250
seed = 42

datasets_to_plot = {
    name: random_subset(sequences, n_samples=n_samples, seed=seed)
    if len(sequences) > n_samples
    else sequences
    for name, sequences in datasets.items()
}
```

```python
physico_chemical_embedder = PropertyEmbedding(
    predictors=[
        MinMaxNorm(Gravy()),
        MinMaxNorm(MolecularWeight()),
        MinMaxNorm(Charge()),
    ]
)
```

```python
colors = ["red", "green", "blue", "orange"]
```

```python
embeddings = {
    name: physico_chemical_embedder(sequences)
    for name, sequences in datasets_to_plot.items()
}
```

Let's create a PCA, T-SNE and UMAP plot.


```python
plot_pca(embeddings.values(), embeddings.keys(), colors)
```

```python
plot_tsne(embeddings.values(), embeddings.keys(), colors)
```

```python
plot_umap(embeddings.values(), embeddings.keys(), colors)
```
